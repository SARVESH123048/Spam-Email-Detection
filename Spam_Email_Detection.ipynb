```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
import string

# Download stopwords if you haven't already
nltk.download('stopwords')
from nltk.corpus import stopwords

# Load the data
data = pd.read_csv('mail_data.csv')

# Replace 'Message' with the actual column name containing email text
text_column_name = 'Message'  # Replace with your actual column name

# Preprocess the data
data[text_column_name] = data[text_column_name].str.lower()
data[text_column_name] = data[text_column_name].fillna('')  # Handle missing values by replacing with empty string
data[text_column_name] = data[text_column_name].apply(lambda text: text.translate(str.maketrans('', '', string.punctuation)))  # Remove punctuation
stop_words = set(stopwords.words('english'))
data[text_column_name] = data[text_column_name].apply(lambda text: ' '.join([word for word in text.split() if word not in stop_words]))

# Split the data
X_train, X_test, y_train, y_test = train_test_split(data[text_column_name], data['Category'], test_size=0.2, random_state=42)

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Now you have X_train_vec, X_test_vec, y_train, y_test ready for model training
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()  # Create a Multinomial Naive Bayes model
model.fit(X_train_vec, y_train)  # Train the model
```

```python
y_pred = model.predict(X_test_vec)  # Predict on the test data
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy
print(f"Accuracy: {accuracy}")  # Print the accuracy
```
